```python
#Make a empty dictionary that maps the column names year and total_births to empty lists:

births_dict = {'year':[], 'total_births':[]}
```

```python
#Loop over the years from 1880 to 2020

for year in range(1880, 2021):
    births = parse_dataset(year)["frequency"].sum()
    births_dict["total_births"].append(births)
    births_dict["year"].append(year)
    total_births = pd.DataFrame(births_dict)
total_births
```

```python
#use pandas to make a quick plot of your results

total_births.plot(x='year', y='total_births', marker='o', linestyle='-', color='skyblue')
plt.title('Total Births Each Year (1880-2020)')
plt.xlabel('Year')
plt.ylabel('Total Births')
plt.grid(True)
plt.show()
```

```python
#How has the number of unique names given to newborn children varied over time?

unique_names_dict = {'year':[], 'unique_names':[]}

for year in range(1880, 2021):
    df_temp = parse_dataset(year)
    unique_names_count = df_temp['name'].nunique()
    unique_names_dict['year'].append(year)
    unique_names_dict['unique_names'].append(unique_names_count)
   
unique_names_df = pd.DataFrame(unique_names_dict)
unique_names_df
```

```python
# Visualize the data, use the years on the x-axis and the counts on the y-axis.

plt.figure(figsize=(15, 10))
plt.bar(unique_names_df['year'], unique_names_df['unique_names'], color='skyblue')
plt.title('Number of Unique Names Given to Newborns Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Unique Names')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Plot the distribution of the length of each name for a given year

def parse_dataset(year):
    path = f'../data/names/yob{year}.txt'
    df = pd.read_csv(path, names=['name', 'gender', 'frequency'])
    df['year'] = year
    return df


selected_year = 2000

selected_dataset = parse_dataset(selected_year)


selected_dataset['name_length'] = selected_dataset['name'].apply(len)


print(selected_dataset.head())


plt.figure(figsize=(10, 6))
plt.hist(selected_dataset['name_length'], bins=range(1, max(selected_dataset['name_length'])), color='pink',)
plt.title(f'Distribution of Name Lengths in {selected_year}')
plt.xlabel('Name Length')
plt.ylabel('Frequency')
plt.show()
```

```python
#Plot the average length of names over time!
average_lengths = []
for year in range(1880, 2021):
    df_year = parse_dataset(year)
    average_length = df_year['name'].apply(len).mean()
    average_lengths.append(average_length)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1880, 2021), average_lengths, marker='o')
plt.title('Average Length of Names Over Time (1880-2020)')
plt.xlabel('Year')
plt.ylabel('Average Name Length')
plt.grid(True)
plt.show()
```

```python
#Analysis of the name 'Arya' before vs After Game of Thrones:
def find_name_frequency(name, start_year, end_year):
    name_frequency = []

    for year in range(1880, 2021+1):
        dataset = parse_dataset(year)
        name_freq = dataset[dataset['name'] == name]['frequency'].sum()
        name_frequency.append({'year': year, 'frequency': name_freq})

    return pd.DataFrame(name_frequency)

Arya_frequency_1880_2011 =find_name_frequency('Arya', 1880, 2011)
Arya_frequency_2011_2021 =find_name_frequency('Arya', 2011, 2021)
```
